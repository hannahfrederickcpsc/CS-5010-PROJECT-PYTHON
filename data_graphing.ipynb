```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# convert our csv file to a data frame
voting_data = pd.read_csv("complete_cleaned.csv")
voting_data
```

```python
pres_years = voting_data.loc[voting_data['Year']%4==0]
gov_years = voting_data.loc[voting_data['Year']%4==1]

year_total_registered = sns.scatterplot(x="Year", y="Total Registered", data=voting_data).set_title('Virginia Voters Registered Through the Years')
plt.savefig('plots/year_total_registered.pdf')
plt.show()
```

```python
sns.scatterplot(x="Year", y="Total Registered", data=voting_data)
sns.scatterplot(x="Year", y="Total Registered", data=pres_years)
year_total_registered_highlighted = sns.scatterplot(x="Year", y="Total Registered", data=gov_years).set_title('Virginia Voters Registered with Highlighted Election Years')
plt.legend(labels=['Other Elections', 'Presidential Elections', 'Gubernatorial Elections'])
plt.savefig('plots/year_total_registered_highlighted.pdf')
plt.show()
```

```python
year_total_voted = sns.scatterplot(x="Year", y="Total Voting", data=voting_data).set_title('Virginia Voters Voting Through the Years')
plt.savefig('plots/year_total_voted.pdf')
plt.show()
```

```python
sns.scatterplot(x="Year", y="Total Voting", data=voting_data)
sns.scatterplot(x="Year", y="Total Voting", data=pres_years)
year_total_voted_highlighted = sns.scatterplot(x="Year", y="Total Voting", data=gov_years).set_title('Virginia Voters Voting with Highlighted Election Years')
plt.legend(labels=['Other Elections', 'Presidential Elections', 'Gubernatorial Elections'])
plt.savefig('plots/year_total_voted_highlighted.pdf')
plt.show()
```

```python
year_total_absentee = sns.scatterplot(x="Year", y="Voting Absentee (Included in Total Voting)", data=voting_data).set_title('Virginia Voters Voting Absentee Through the Years')
plt.savefig('plots/year_total_absentee.pdf')
plt.show()
```

```python
sns.scatterplot(x="Year", y="Voting Absentee (Included in Total Voting)", data=voting_data)
sns.scatterplot(x="Year", y="Voting Absentee (Included in Total Voting)", data=pres_years)
year_total_absentee_highlighted = sns.scatterplot(x="Year", y="Voting Absentee (Included in Total Voting)", data=gov_years).set_title('Virginia Voters Voting Absentee with Highlighted Election Years')
plt.legend(labels=['Other Elections', 'Presidential Elections', 'Gubernatorial Elections'])
plt.savefig('plots/year_total_absentee_highlighted.pdf')
plt.show()
```

```python
year_voter_turnout = sns.scatterplot(x="Year", y="Turnout (% Voting of Total Registered)", data=voting_data).set_title('Virginia Voter Turnout Through the Years')
plt.savefig('plots/year_voter_turnout.pdf')
plt.show()
```

```python
sns.scatterplot(x="Year", y="Turnout (% Voting of Total Registered)", data=voting_data)
sns.scatterplot(x="Year", y="Turnout (% Voting of Total Registered)", data=pres_years)
year_voter_turnout_highlighted = sns.scatterplot(x="Year", y="Turnout (% Voting of Total Registered)", data=gov_years).set_title('Virginia Voter Turnout with Highlighted Election Years')
plt.legend(labels=['Other Elections', 'Presidential Elections', 'Gubernatorial Elections'])
plt.savefig('plots/year_voter_turnout_highlighted.pdf')
plt.show()
```

```python
year_total_registered_party = sns.scatterplot(x="Year", y="Total Registered", hue="Party", data=voting_data).set_title('Total Registered Virginia Voters Through the Years By Winning Party')
plt.savefig('plots/year_total_registered_party.pdf')
plt.show()
```

```python
post_motor_voter = voting_data.loc[voting_data['Year'] >= 1996]

year_total_registered = sns.scatterplot(x="Year", y="Total Registered", data=voting_data).set_title('Virginia Voters Registered Through the Years')
year_total_registered_motor = sns.scatterplot(x="Year", y="Total Registered", data=post_motor_voter).set_title('Virginia Voters Registered with Highlighted Motor Voter Years')
plt.legend(labels=['Pre-Motor Voter Act', 'Post-Motor Voter Act'])
plt.savefig('plots/year_total_registered_motor.pdf')
plt.show()
```

```python
year_total_voted = sns.scatterplot(x="Year", y="Total Voting", data=voting_data).set_title('Virginia Voters Voting Through the Years')
year_total_voted_motor = sns.scatterplot(x="Year", y="Total Voting", data=post_motor_voter).set_title('Virginia Voters Voting with Highlighted Motor Voter Years')
plt.legend(labels=['Pre-Motor Voter Act', 'Post-Motor Voter Act'])
plt.savefig('plots/year_total_voted_motor.pdf')
plt.show()
```
