```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# convert our csv file to a data frame
voting_data = pd.read_csv("complete_cleaned.csv")
voting_data.head()
```

```python
#make and save a boxplot of the total registered voters by which party won
total_registered_box = sns.boxplot(x="Party", y="Total Registered", data=voting_data).set_title('Virginia Voters Registered by Which Party Won')
plt.savefig('plots/total_registered_box.pdf')
plt.show()
#note the average and the variance of the boxplots
```

```python
#make and save a boxplot of the total actual voters by which party won
total_voted_box = sns.boxplot(x="Party", y="Total Voting", data=voting_data).set_title('Virginia Voters Voting by Which Party Won')
plt.savefig('plots/total_voted_box.pdf')
plt.show()
#note the average and the variance of the boxplots
```

```python
#make and save a boxplot of the total absentee voters by which party won
total_absentee_box = sns.boxplot(x="Party", y="Voting Absentee (Included in Total Voting)", data=voting_data).set_title('Virginia Voters Voting Absentee by Which Party Won')
plt.savefig('plots/total_absentee_box.pdf')
plt.show()
#note the average and the variance of the boxplots
```

```python
#make and save a boxplot of the percent voting of registered voters by which party won
total_turnout_box = sns.boxplot(x="Party", y="Turnout (% Voting of Total Registered)", data=voting_data).set_title('Virginia Voter Turnout by Which Party Won')
plt.savefig('plots/total_turnout_box.pdf')
plt.show()
#note the average and the variance of the boxplots
```

```python
#make a dataframe with only presidential elections
pres_years = voting_data.loc[voting_data['Year']%4==0]
#make a dataframe with only gubernatorial elections
gov_years = voting_data.loc[voting_data['Year']%4==1]
#make a dataframe with all the other elections
other_years = voting_data.loc[(voting_data['Year']%4==2) | (voting_data['Year']%4==3)]
#make and save a line plot of the total registered voters through the years
year_total_registered = plt.plot("Year", "Total Registered", data=voting_data, marker='o') #make the plot of all elections
plt.title('Registered Virginia Voters Through the Years') #add the title
plt.xlabel('Year') #add the x-axis label
plt.ylabel('Number of Registered Virginia Voters') #add the y-axis label
#year_total_registered = sns.scatterplot(x="Year", y="Total Registered", data=voting_data).set_title('Virginia Voters Registered Through the Years')
plt.savefig('plots/year_total_registered.pdf')
plt.show()
```
