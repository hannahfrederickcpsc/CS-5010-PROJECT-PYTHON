```python
gov_years_data = voting_data.loc[voting_data['Year']%4==1]
gov_years_data["Change in Total Voting"] = (gov_years_data["Total Voting"] / gov_years_data["Total Voting"].shift(-1)) -1

least_perc_change = abs(gov_years_data["Change in Total Voting"]).min()
least_perc_change_years = gov_years_data[gov_years_data["Change in Total Voting"] == least_perc_change]
least_perc_change_years
```

```python
# begin andre's queries
# evaluate the mean turnouts in presidential and non-presidential years
pres_years_data = voting_data.loc[voting_data['Year']%4==0]
pres_turnout_mean = pres_years_data['Turnout (% Voting of Total Registered)'].mean()
pres_turnout_mean
```

```python
nonpres_years_data = voting_data.loc[voting_data['Year']%4!=0]
nonpres_turnout_mean = nonpres_years_data['Turnout (% Voting of Total Registered)'].mean()
nonpres_turnout_mean
```

```python
diff_percent = round(abs((pres_turnout_mean - nonpres_turnout_mean))*100,3)
diff_percent
```

```python
# two tailed t-test to see if this is a significant difference in means
stats.ttest_ind(pres_years_data['Turnout (% Voting of Total Registered)'], nonpres_years_data['Turnout (% Voting of Total Registered)'], equal_var=False)
print('The p-value of the two-tailed t-test is very low, indicating that there is a significant difference in the mean turnout in presidential election years versus non-presidential election years.')
```

```python
# establish dict of presidential election years and winners
pres_winners = {'1976':'Jimmy Carter', '1980':'Ronald W. Reagan', '1984':'Ronald W. Reagan',
                '1988':'George H. Bush', '1992': 'Bill Clinton', '1996': 'Bill Clinton',
                '2000': 'Bush and Cheney', '2004':'Bush and Cheney', '2008': 'Barack Obama',
                '2012': 'Barack Obama', '2016': 'Donald Trump'}
```

```python
## define function to:
## take an election year
## check data against pres_winners dict
## return a boolean
def va_elect(year):
    return pres_winners[year] == pres_years_data.loc[pres_years_data.Year == int(year), 'Winner (VA)'].values[0]
```

```python
check_year = input('Please enter a presidential election year: ')
try:
    print('Did Virginia results align with the nation in this election? ' + str(va_elect(check_year)))
except KeyError:
    print('You did not enter a presidential election year.')
```
